```python
import subprocess

def check_video_file(input_file):
    # 构建FFmpeg命令
    ffmpeg_command = [
        'ffmpeg',
        '-i', input_file,
        '-c', 'copy',
        '-f', 'null', 
        '-'
    ]
    
    # 执行FFmpeg命令，并捕获输出
    try:
        result = subprocess.run(ffmpeg_command, check=True, stdout=subprocess.PIPE, stderr=subprocess.PIPE)
        output = result.stderr.decode('utf-8', errors='ignore')
        print("FFmpeg output:\n", output)
        
        # 解析输出，检查潜在问题
        if "error" in output.lower():
            print("Errors detected in the video file:")
            for line in output.splitlines():
                if "error" in line.lower():
                    print(line)
        else:
            print("No errors detected in the video file.")
        
    except subprocess.CalledProcessError as e:
        output = e.stderr.decode('utf-8', errors='ignore')
        print("Failed to analyze video file:\n", output)

# 示例调用
input_file = r'D:\小狗视频\2月\20240202kt\020500.mp4'
check_video_file(input_file)
```

```python
def extract_frames_with_vsync(input_file, frame_dir, vsync='0'):
    if not os.path.exists(frame_dir):
        os.makedirs(frame_dir)

    ffmpeg_command = [
        'ffmpeg',
        '-vsync', vsync,  # 设置视频同步模式
        '-i', input_file,
        os.path.join(frame_dir, 'frame_%04d.png')
    ]
    
    try:
        result = subprocess.run(ffmpeg_command, check=True, stdout=subprocess.PIPE, stderr=subprocess.PIPE)
        print(f"Extracted frames to {frame_dir} with vsync={vsync}")
        print("FFmpeg output:\n", result.stdout.decode('utf-8', errors='ignore'))
        print("FFmpeg errors:\n", result.stderr.decode('utf-8', errors='ignore'))
    except subprocess.CalledProcessError as e:
        print(f"Failed to extract frames with vsync={vsync}: {e}")
        print("FFmpeg errors:\n", e.stderr.decode('utf-8', errors='ignore'))

input_file = r'D:\小狗视频\2月\20240202kt\020500.mp4'
base_output_dir = r"D:\小狗视频\2月\20240202kt_output"

vsync_values = ['0', '1', '2', 'passthrough']

for vsync in vsync_values:
    frame_dir = os.path.join(base_output_dir, f"frames_vsync_{vsync}")
    extract_frames_with_vsync(input_file, frame_dir, vsync)
```

```python
def extract_frames_with_decoder(input_file, frame_dir, decoder):
    if not os.path.exists(frame_dir):
        os.makedirs(frame_dir)

    ffmpeg_command = [
        'ffmpeg',
        '-c:v', decoder,  # 指定解码器
        '-i', input_file,
        os.path.join(frame_dir, 'frame_%04d.png')
    ]
    
    try:
        result = subprocess.run(ffmpeg_command, check=True, stdout=subprocess.PIPE, stderr=subprocess.PIPE)
        print(f"Extracted frames to {frame_dir} with decoder={decoder}")
        print("FFmpeg output:\n", result.stdout.decode('utf-8', errors='ignore'))
        print("FFmpeg errors:\n", result.stderr.decode('utf-8', errors='ignore'))
    except subprocess.CalledProcessError as e:
        print(f"Failed to extract frames with decoder={decoder}: {e}")
        print("FFmpeg errors:\n", e.stderr.decode('utf-8', errors='ignore'))

input_file = r'D:\小狗视频\2月\20240202kt\020500.mp4'
base_output_dir = r"D:\小狗视频\2月\20240202kt_output"

decoders = ['h264', 'libopenh264', 'h264_cuvid']

for decoder in decoders:
    frame_dir = os.path.join(base_output_dir, f"frames_decoder_{decoder}")
    extract_frames_with_decoder(input_file, frame_dir, decoder)
```

```python
import os
import subprocess

def extract_frames_with_specific_stream(input_file, frame_dir, stream_index):
    if not os.path.exists(frame_dir):
        os.makedirs(frame_dir)

    # 使用 FFmpeg 提取指定流的所有帧
    ffmpeg_command = [
        'ffmpeg',
        '-i', input_file,
        '-map', f'0:{stream_index}',  # 指定要处理的流
        os.path.join(frame_dir, 'frame_%04d.png')
    ]
    
    try:
        result = subprocess.run(ffmpeg_command, check=True, stdout=subprocess.PIPE, stderr=subprocess.PIPE)
        print(f"Extracted frames to {frame_dir} from stream {stream_index}")
        print("FFmpeg output:\n", result.stdout.decode('utf-8', errors='ignore'))
        print("FFmpeg errors:\n", result.stderr.decode('utf-8', errors='ignore'))
    except subprocess.CalledProcessError as e:
        print(f"Failed to extract frames from stream {stream_index}: {e}")
        print("FFmpeg errors:\n", e.stderr.decode('utf-8', errors='ignore'))

# 示例调用
input_file = r'D:\小狗视频\2月\20240202kt\020500.mp4'
base_output_dir = r"D:\小狗视频\2月\20240202kt_output"

# 选择流1（1920x1080）进行处理
stream_index = 0
frame_dir = os.path.join(base_output_dir, f"frames_stream_{stream_index}")
extract_frames_with_specific_stream(input_file, frame_dir, stream_index)
```
