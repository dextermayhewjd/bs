```python
import os
import json
from collections import defaultdict
```

```python

# 设定目录路径
directory_path = "D:\\小狗视频"

# 用于存储所有行为标签及其对应文件的信息
labels_info = defaultdict(list)

# 遍历指定目录及其子目录中的所有文件
for subdir, dirs, files in os.walk(directory_path):
    for file in files:
        if file.endswith(".json"):
            file_path = os.path.join(subdir, file)
            with open(file_path, 'r', encoding='utf-8') as f:
                data = json.load(f)
                # 提取每个文件中的行为标签
                for key, value in data["metadata"].items():
                    if "av" in value:
                        for label in value["av"].values():
                            # 记录标签及其对应的文件和文件夹信息
                            labels_info[label].append({
                                "folder": os.path.basename(subdir),
                                "file": file
                            })

# 打印统计结果
for label, occurrences in labels_info.items():
    print(f"Label '{label}' occurs in the following files:")
    for occurrence in occurrences:
        print(f"  Folder: {occurrence['folder']}, File: {occurrence['file']}")
    print()
```
