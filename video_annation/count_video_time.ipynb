```python
import subprocess
import json

```

```python
def get_video_duration(video_file):
    # 使用ffprobe获取视频的持续时间
    cmd = ['ffprobe', '-v', 'error', '-show_entries', 'format=duration', '-of', 'default=noprint_wrappers=1:nokey=1', video_file]
    result = subprocess.run(cmd, stdout=subprocess.PIPE, stderr=subprocess.STDOUT, text=True)
    duration = float(result.stdout.strip())
    return duration

# 示例使用
video_file = r'D:\小狗视频\2月\20240201kt\004800.mp4'  # 确保这里的路径和文件名是正确的
duration_in_seconds = get_video_duration(video_file)
print(f"Video duration: {duration_in_seconds} seconds")
```

```python
from moviepy.editor import VideoFileClip
import os

def calculate_video_durations(directory_path):
    total_duration = 0.0
    video_count = 0  # 用于记录视频编号
    for root, dirs, files in os.walk(directory_path):
        for file in files:
            if file.endswith(('.mp4', '.avi', '.mov')):  # 添加你需要处理的视频格式
                try:
                    filepath = os.path.join(root, file)
                    with VideoFileClip(filepath) as video:
                        duration = video.duration
                        video_count += 1  # 处理完一个视频，编号加1
                        print(f"Video #{video_count}: {file}, Duration: {duration:.3f} seconds")
                        total_duration += duration
                except Exception as e:
                    print(f"Error processing file {file}: {str(e)}")
    print(f"Processed {video_count} videos.")
    print(f"Total duration of all videos: {total_duration:.3f} seconds")

directory_path = 'D:\\小狗视频\\2月\\20240201kt'  # 
calculate_video_durations(directory_path)
```
